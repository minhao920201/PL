```python
import json #載入json模組
```

```python
test = open('csvjson.json',encoding="utf8") #開檔
Data = json.load(test) #反序列化(decode)
print(type(Data)) #印出 Data 的格式
```

```python
Data.items() #印出所有 Data 裡面的 key 和 value
```

```python
temp = Data.get('Title') #取得 key = "Title" 的 value 
temp #印出 key = "Title" 的 value
```

```python
dic = {} #紀錄下面取得的不同"標籤"出現次數
for i in temp: #從 list 中一個一個列舉出 dict
    #print(li['標籤'])
    for j in i['標籤']: #列舉出每個 key = "標籤" 的 value
        dic[j] = dic.get(j,0)+1 #檢查"標籤"中的每個字串是否出現在 dic 中。如果沒出現，則加入 dic 裡，並使其 value=1 ; 否則已經出現就將其 value+1
dic #印出不同"標籤"出現次數
```

```python
#查詢1 (出現次數最多的"標籤")
id = 0 #下面 li_key 和 li_value 的索引值
count = 0 #紀錄現在讀取道第n筆的資料
max = 0 #為了找出最常出現的"標籤"
li_key = []
li_value = []
for i in dic:
    li_key.append(i) #將每個標籤名稱加入 list 中
    li_value.append(dic[i]) #將每個標籤出現過的次數加入 list 中
    if dic[i]>max: #確認是否為最常出現的"標籤"
        max = dic[i]
        id = count
    count+=1
print("最常出現的標籤 :",li_key[id],", 出現",li_value[id],"次")
```

```python
#查詢二 (自己輸入想查詢的資料)
#以下適用於查詢"標籤"
ask = input() #使用者輸入任一標籤
for i in temp: #利用迴圈找出所有相關的文章
    for j in i['標籤']:
        if ask == j:
            print(i)
```

```python
#排序1 (由高到低排序"標籤出現的次數")
li_value.sort(reverse = True) #將"標籤出現次數"由大到小排序
li_key_sorted = [] #存放排序好的"標籤"
length = len(li_key) #"標籤"總數

#利用迴圈排序"標籤"
for i in range(length):
    for j in dic:
        if li_value[i] == dic[j]:
            li_key_sorted.append(j)
            
#利用迴圈印出排序後的結果
for i in range(length):
    print(li_key_sorted[i],":",li_value[i])
```
