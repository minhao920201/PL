```python
import json #載入json模組
test = open('JSONcrawler.json',encoding="utf8") #開檔
Data = json.load(test) #反序列化(decode)
#把蒐集到的資料印出來
temp = Data.get('Article') #取得 key = "Title" 的 value 
temp #印出 key = "Title" 的 value
```

```python
import jieba
import jieba.analyse
jieba.analyse.set_stop_words('stopwords.txt')
index = 0
li_jieba = []
for documents in temp:
    seg_list = jieba.analyse.extract_tags(documents['內文'], 150)
    li_jieba.append(seg_list)
```

```python
li_jieba
```

```python
#印出不同"標籤"出現次數
dic = {} #紀錄下面取得的不同"標籤"出現次數
for i in temp: #從 list 中一個一個列舉出 dict
    #print(li['標籤'])
    for j in i['標籤']: #列舉出每個 key = "標籤" 的 value
        dic[j] = dic.get(j,0)+1 #檢查"標籤"中的每個字串是否出現在 dic 中。如果沒出現，則加入 dic 裡，並使其 value=1 ; 否則已經出現就將其 value+1
dic #利用dic印出不同"標籤"出現次數
```

```python
#利用迴圈顯示出標籤次數大於10的"標籤"
for i in dic:
    if dic[i]>10:
        print(i,":",dic[i])
```

# 製作長條圖

```python
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties

#把dic中的key和value抽取出來，以便下方製作圖使用
li_key = []
li_value = []
for i in dic:
    if dic[i]>10:
        li_key.append(i)
        li_value.append(dic[i])

#把網路上下載的字體加入
myfont = FontProperties(fname=r'./NotoSerifTC-Black.otf')

#設定圖片輸出的資訊
plt.figure(figsize=(20,20))
plt.title("Top Question")
plt.xlabel("Tag")
x = range(len(li_key))
plt.xticks(rotation=30,ha = "right")
plt.ylabel("Frequency")
plt.bar(li_key,li_value)
plt.xticks(x,li_key,fontproperties=myfont)
plt.savefig("Top Questions Chart")
plt.show()
```

# 製作文字雲

```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud

import numpy as np
mask = np.array(Image.open("it.jpg"))
font = "C:\Windows\Fonts\MSJH.TTC"

wordcloud = WordCloud(background_color = "white", font_path=font, mask = mask)
wordcloud.generate_from_frequencies(dic)

plt.figure(figsize=(20,20))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("Word Cloud")
plt.show()
```
