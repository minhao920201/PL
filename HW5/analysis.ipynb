```python
import json #載入json模組
test = open('JSONcrawler.json',encoding="utf8") #開檔
Data = json.load(test) #反序列化(decode)
#把蒐集到的資料印出來
temp = Data.get('Article') #取得 key = "Title" 的 value 
temp #印出 key = "Title" 的 value
```

```python
import pandas as pd
df = pd.DataFrame(temp)
df
```

```python
df['seg'] = ""
df.head()
```

```python
import jieba
import jieba.analyse
import re
jieba.analyse.set_stop_words('stopwords.txt')
index = 0
li_jieba = []
for documents in df['內文']:
    #documents = re.sub(r"EOL","",documents)
    documents = re.sub(r"[a-zA-Z0-9]","",documents)
    seg_list = jieba.analyse.extract_tags(documents, 150)
    li_jieba.append(seg_list)
    #seg_list = jieba.lcut(documents)
    df['seg'].iloc[index] = seg_list
    index += 1
li_jieba
```

```python
#印出不同"標籤"出現次數
dic = {} #紀錄下面取得的不同"標籤"出現次數
for i in temp: #從 list 中一個一個列舉出 dict
    #print(li['標籤'])
    for j in i['標籤']: #列舉出每個 key = "標籤" 的 value
        dic[j] = dic.get(j,0)+1 #檢查"標籤"中的每個字串是否出現在 dic 中。如果沒出現，則加入 dic 裡，並使其 value=1 ; 否則已經出現就將其 value+1
dic #利用dic印出不同"標籤"出現次數
```

```python
#利用迴圈顯示出標籤次數大於10的"標籤"
for i in dic:
    if dic[i]>10:
        print(i,":",dic[i])
```

# 依照出現次數的多寡進行排序，以對照下方產生的文字雲正確性

```python
li_key = []
li_value = []
for i in dic:
    li_key.append(i) #將每個標籤名稱加入 list 中
    li_value.append(dic[i]) #將每個標籤出現過的次數加入 list 中
#排序 (由高到低排序"標籤出現的次數")
li_value.sort(reverse = True) #將"標籤出現次數"由大到小排序
li_key_sorted = [] #存放排序好的"標籤"
length = len(li_key) #"標籤"總數

#利用迴圈排序"標籤"
for i in range(length):
    for j in dic:
        if li_value[i] == dic[j]:
            li_key_sorted.append(j)

dic_sort = {}
#利用迴圈印出排序後的結果
for i in range(length):
    dic_sort[li_key_sorted[i]] = li_value[i]
    print(li_key_sorted[i],":",li_value[i])
```

```python
import random
li_rand = []
for i in range(10):
    li_rand.append(random.randrange(314))
li_rand
```

```python
dic_rand = {}
count = 0
for i in li_rand:
    for key in dic_sort:
        if dic_sort[key]==i:
            dic_rand[key] = dic_sort[key]
            count += 1
dic_rand
```

```python
res = pd.DataFrame(columns = ['tag','seg'],index = range(count))
cnt = 0
for tag in dic_rand:
    res['tag'][cnt] = tag
    cnt += 1
res
```

```python
count = 0
for tag in dic_rand:
    id = 0
    li_id = []
    for seg in df['標籤']:
        if tag in seg:
            li_id.append(id)
        id += 1
    big_list = []
    for i in li_id:
        for seg in df['seg'][i]:
            big_list.append(seg)
    res['seg'][count] = big_list
    count += 1
res
```
