```python
import pandas as pd #載入pandas套件
from IPython.display import display #載入Ipython套件
data = pd.read_csv("IT_data.csv") #讀取csv檔
data.head() #印出前五筆data
```

```python
#總共有七天的資料量，將近期三天的資料作為一個集合
first_date = data[data["詢問日期"]=="9月24日"] #找出日期為9/24的data
second_date = data[data["詢問日期"]=="9月23日"] #找出日期為9/23的data
third_data = data[data["詢問日期"]=="9月22日"] #找出日期為9/22的data
first_three_date = pd.concat([first_date,second_date,third_data],axis=0) #將三個data整合在一起
display(first_three_date) #如果使用 print 印格式會跑掉，因為jupyter 會轉成 html table顯示，故選擇用 display 顯示
```

```python
#資料中的"標籤"有許多種，以 python 當作一個集合
py = data[data["標籤"]=="python"] #找出標籤名稱為python的data
display(py)
```

```python
#交集 (將有python標籤的data跟近期三天的data做交集)
intersection = pd.merge(first_three_date,py,how="inner") 
display(intersection)
```

```python
#聯集 (將有python標籤的data跟近期三天的data做聯集)
union = pd.merge(first_three_date,py,how="outer")
display(union)
```

```python
#差集1 (py - first_three_date)
difference1 = pd.concat([py,first_three_date,first_three_date]).drop_duplicates(keep=False)
display(difference1)
```

```python
#差集2 (first_three_date - py)
difference2 = pd.concat([first_three_date,py,py]).drop_duplicates(keep=False)
display(difference2)
```

```python
#驗證差集二 index = 44 和 index = 100 的資料不包含在 data 中
difference2["文章標題"][44]
```

```python
#驗證差集二 index = 44 和 index = 100 的資料不包含在 data 中
difference2["文章標題"][100]
```

```python
#對稱差集
Symmetrical_difference = pd.merge(difference1,difference2,how="outer") #將差集1和差集2做聯集
display(Symmetrical_difference)
```
